Time plot?!



````python
pupil_left = df['Pupil diameter left']
````

````python
figure(figsize=(21, 7))
plt.plot(pupil_left, label="Raw Pupil Diameter")
plt.show()
````

````python
figure(figsize=(21, 7))
pupil_left = pupil_left[~np.isnan(pupil_left)]
plt.plot(pupil_left, label="Raw Pupil Diameter")

# applying the butterworth filter (low pass filter)
cutoff = .001
b, a = scipy.signal.butter(3, cutoff)
filtered = scipy.signal.filtfilt(b, a, pupil_left)
mean_value_smoothed = np.median(filtered)
plt.plot(filtered, label="Smoothed Pupil Diameter")

plt.legend()
plt.title("Raw vs. Smoothed Pupil Diameter")
#         plt.ylim([1.5, 4.5])
plt.xlabel('Stimulus Presentation Time [ms]')
plt.ylabel('Pupil Diameter [nm]')

plt.show()
````
