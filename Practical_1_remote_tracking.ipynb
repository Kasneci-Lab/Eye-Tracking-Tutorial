# Eye-based Human-Computer Interaction
# Practical 1: Remote eye-tracking

In this first session we will familiarize ourselves with remote eye-tracking systems.
Each group will perform a recording on one specific eye-tracking device.

Don't be frustrated if other groups make faster progress than you do! It might be entirely due to an easier to use system. That is intended and part of the experience. It's more about learning what to look out for when using an eye-tracker than getting the perfect recording done.

After you performed your own recording, we dissolve the groups and participate in each other's experiments. That's an excellent chance to get to know all the other systems. Please take your time and ask yout fellow students a lot of questions. Explain to the others what's cool / annoying about your device.

## Task 2: Yarbus' recording
In a famous experiment Yarbus was able to show that we view paintings in a different way when we are given different tasks (https://wwws.uni-weimar.de/kunst-und-gestaltung/wiki/images/Unexpected_visitor.pdf). Our intentions and internal thoughts change how our eyes move. Whether he asked "How long was the visitor away" or "How wealthy is the family", different objects are fixated. We will recreate his experiment:




image.png

Use the stimulus presentation software available for your Eye-Tracker (e.g., Tobii Studio, Ogama, ...) to recreate the above stimulus order (Fixations crosses followed by Ilja Repin's "Unexpected Visitor", available here: https://upload.wikimedia.org/wikipedia/commons/1/16/Ilya_Repin_Unexpected_visitors.jpg). The repetition is intended for two different questions being asked.

**Part 1:**
- Find out how your stimulus presentation software (or your workaround in case of a lack of such) can present the images and the fixation cross.
- Choose two of Yarbus' original tasks that you expect to deliver different scanpaths.
- Think about how long you present each stimulus / mask and how to proceed to the next image.
- Find out how to calibrate the eye-tracker and how to check/quantify whether a calibration was good (if possible).
- Record yourselves.

**Part 2:**

- Think about how to instruct your subjects to perform the experiment correctly.
- Record with subjects from the other groups as well. This is intended as an exchange between groups, not as a massive data acquisition session. Take your time to explain your eye-tracking device and software to them, possibly highlighting what you liked about it and what problems you ran into. There is no minimum number of subejcts you need to record! Please focus on teaching and learning with the others.
- Participate in the other groups' experiments and please do ask questions, e.g. about their devices, experimental choices, software, live gaze visualization demo, ...

## Task 2: Export your data
For interoperability between devices and software most vendors allow to export gaze data to a CSV/text file (of some sort and structure).
Export your data and write a minimal Readme file on how to interpret your data. We will exchange data between groups for the next sessions. So make sure the other groups can easily parse and correctly interpret your data, specifically:
- the timestamp (which unit?)
- the x and y coordinate of gaze on screen, for both eyes if available
- whether a gaze sample is valid or not
- what is currently displayed on the screen (fixation cross / painting)
- where on the screen the painting/cross is displayed (e.g. screen resolution / origin and pixel size in case of non-full-screen)

- **Please hand in the data to us!**

##Task 3: Analyze your data with the ready-made software

What kind of analysis can you perform with the software by your vendor (Tobii / Ogama / ...)? We will have a super short 5 minute presentation where each group can present their data and a very first analysis of what you think is important / remarkable / strange about it.

Some inspiration on the possible content (neither a complete list nor a list of requirements - some of them might not be easily doable with your software):
- How much data did you record?
- Can you say anything about data quality?
- Anything else you want the other groups to know before they start working on your data?
- Were there irregularities during recording that might be important to know when working with the data? (Please write those down as well)
- Can we visualize the data?
- Can we somehow even compare the two questions asked?
